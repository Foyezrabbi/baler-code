```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import chromedriver_autoinstaller
import csv
import time
```

```python
def driver_define():
    print('Chromedriver Installing')
    driver_path = chromedriver_autoinstaller.install()
    
    print('Chrome Browser Opening')
    options = Options()
    options.add_experimental_option('excludeSwitches', ['enable-logging'])
    s = Service(driver_path)
    driver = webdriver.Chrome(service=s, options =options)
    return driver
```

```python
driver = driver_define()
```

```python
i = 1
with open("untitled.txt", "r") as file:
    for url in file.readlines():
        driver.get(url)
        
        try:
            title = driver.find_element(By.ID, 'h1PartTitle').text
            print(title)
        except Exception:
            title = "None"
            
        try:
            description = driver.find_element(By.CLASS_NAME, 'feaures-benefits-box').text
            print("description")
        except Exception:
            description = "None"
        
        try: 
            image_src = driver.find_element(By.XPATH, '//*[@id="main-content"]/div[2]/div[2]/div[1]/div[1]/img').get_attribute("src")
            print(image_src)
        except Exception:
            image_src = "None"
            
        try:            
             bar_name1 = driver.find_element(By.XPATH, "//*[@id='main-content']/div[2]/div[2]/div[4]/ul/li[1]")
             bar_name2 = driver.find_element(By.XPATH, "//*[@id='main-content']/div[2]/div[2]/div[4]/ul/li[2]")
            
             if "Technical Resources" in bar_name1.text: 
                bar_name1.click()
             elif "Technical Resources" in bar_name2.text:
                bar_name2.click()
             else:
                print("error")
             time.sleep(2)
            
        except Exception:
            pass
            
        try:
            pdf_list = []
            pdf_links = driver.find_elements(By.XPATH, '//*[@id="tabTechRes"]//td/a')
            for link in pdf_links:
                pdf_list.append(link.get_attribute("href"))
            csv_pdf = ", ".join(pdf_list)
            print(csv_pdf)
        except Exception:
            csv_pdf = "None"
        
        print("------------------ ", i)
        with open('info.csv', 'a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow([i, title, description, image_src, csv_pdf])
            i += 1
            
    
            
```
